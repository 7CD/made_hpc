```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()

G.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 4), (4, 1)])

nx.draw_networkx(G, with_label = True, node_color ='green') 
```

```python
G = nx.Graph()

G.add_edges_from([(0, 1) (1, 2), (2, 0)])#, (3, 2) (3, 1), (4, 5), (5, 2)])

nx.draw_networkx(G, with_label = True, node_color ='green') 
```
